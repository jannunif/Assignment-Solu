```python
#ans01 import re

def check_password(password):
    # Define regex patterns for each requirement
    pattern_uppercase = r'[A-Z].*[A-Z]'
    pattern_lowercase = r'[a-z].*[a-z]'
    pattern_number = r'\d'
    pattern_special = r'[!@#$%^&*()]'
    
    # Check if password meets all requirements
    if (len(password) == 10 and 
        re.match(pattern_uppercase, password) and 
        re.match(pattern_lowercase, password) and 
        re.findall(pattern_number, password) and 
        len(re.findall(pattern_special, password)) >= 3):
        return "Valid Password"
    else:
        return "Invalid Password"
```
